```python
import matplotlib.pyplot as plt
from IPython.display import Image
import numpy as np
import scipy
from scipy.integrate import quad, dblquad, tplquad
from scipy.integrate import odeint, ode
%matplotlib inline
```

```python

def f(y, t, zeta, w0):
    """ первая часть уравнения"""
    x, p = y[0], y[1]
    dx = p
    dp = -2 *zeta*w0*p-w0**2*x
    return [dx, dp]

y0 = [1.0, 0.0]

t = np.linspace(0, 10, 1000)
w0 = 2*np.pi*1.0

y1 = odeint(f, y0, t, args=(0.0, w0))
y2 = odeint(f, y0, t, args=(0.2, w0))
y3 = odeint(f, y0, t, args=(1.0, w0))
y4 = odeint(f, y0, t, args=(5.0, w0))


fig, ax = plt.subplots()
ax.plot(t, y1[:, 0], 'k', label='незатухающий', linewidth=0.25)
ax.plot(t, y1[:, 0], 'k', label='демпфирование')
ax.plot(t, y1[:, 0], 'k', label='критическое затухание')
ax.plot(t, y1[:, 0], 'k', label='чрезмерное демпфирование')
ax.legend();
```
